```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
image = cv2.imread('./examples/images/02/1.png')
```

```python
# 平移变换
M_translation = np.float32([[1, 0, 100], [0, 1, 50], [0, 0, 1]])
image_translated = cv2.warpPerspective(image, M_translation, (image.shape[1], image.shape[0]))

# 翻转变换
M_flip = np.float32([[-1, 0, image.shape[1]], [0, -1, image.shape[0]], [0, 0, 1]])
image_flipped = cv2.warpPerspective(image, M_flip, (image.shape[1], image.shape[0]))

# 缩放变换
M_scale = np.float32([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 1]])
image_scaled = cv2.warpPerspective(image, M_scale, (image.shape[1], image.shape[0]))

# 旋转变换
theta = np.radians(30)  # 旋转30度
M_rotate = np.float32([[np.cos(theta), -np.sin(theta), 0], [np.sin(theta), np.cos(theta), 0], [0, 0, 1]])
image_rotated = cv2.warpPerspective(image, M_rotate, (image.shape[1], image.shape[0]))

# 仿射变换
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])
M_affine = cv2.getAffineTransform(pts1, pts2)
image_affine = cv2.warpAffine(image, M_affine, (image.shape[1], image.shape[0]))
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(231), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(232), plt.imshow(cv2.cvtColor(image_translated, cv2.COLOR_BGR2RGB)), plt.title('Translation')
plt.subplot(233), plt.imshow(cv2.cvtColor(image_flipped, cv2.COLOR_BGR2RGB)), plt.title('Flip')
plt.subplot(234), plt.imshow(cv2.cvtColor(image_scaled, cv2.COLOR_BGR2RGB)), plt.title('Scale')
plt.subplot(235), plt.imshow(cv2.cvtColor(image_rotated, cv2.COLOR_BGR2RGB)), plt.title('Rotation')
plt.subplot(236), plt.imshow(cv2.cvtColor(image_affine, cv2.COLOR_BGR2RGB)), plt.title('Affine Transformation')

plt.show()
```
